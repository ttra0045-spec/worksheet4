## Part 2: Numeric Distributions

Plot distributions for a few numeric variables, such as:

- `total_hours`
- `avg_hourly_pay`
- `weekly_sales`

Then briefly reflect on what you see.


```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.histplot(data=df, x='total_hours', ax=axes[0])
axes[0].set_title('Distribution of Total Hours')

sns.histplot(data=df, x='avg_hourly_pay', ax=axes[1])
axes[1].set_title('Distribution of Average Hourly Pay')

sns.histplot(data=df, x='weekly_sales', ax=axes[2])
axes[2].set_title('Distribution of Weekly Sales')

plt.tight_layout()
plt.show()
```

## Part 3: Composition of a Categorical Variable

Pick one categorical variable, such as `store_type`, `region`, or `promotion_flag`, and show its composition with a suitable plot.


```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.countplot(data=df, x='store_type', ax=axes[0])
axes[0].set_title('Composition of Store Types')

sns.countplot(data=df, x='region', ax=axes[1])
axes[1].set_title('Composition of Regions')

sns.countplot(data=df, x='promotion_flag', ax=axes[2])
axes[2].set_title('Composition of Promotion Flags')

plt.tight_layout()
plt.show()
```

What does this composition tell you? Are some categories dominant or rare?


## Part 4: Scatter Plots and Correlations

Choose a pair of numeric variables and make at least one scatter plot.

Then calculate both:

- Pearson correlation
- Spearman correlation


```python
sns.scatterplot(data=df, x='total_hours', y='avg_hourly_pay')
plt.title('Scatter Plot: Total Hours vs Average Hourly Pay')
plt.show()
```

```python
pearson_corr = df['total_hours'].corr(df['avg_hourly_pay'], method='pearson')
spearman_corr = df['total_hours'].corr(df['avg_hourly_pay'], method='spearman')

print(f"Pearson correlation: {pearson_corr}")
print(f"Spearman correlation: {spearman_corr}")
```

In one or two sentences: do the plots and the correlation coefficients tell the same story?


## Part 5: Change the Scale

Revisit one relationship and change the plotting scale. For example, try a log scale on the y-axis if the raw plot is hard to read.


```python
sns.scatterplot(data=df, x='total_hours', y='avg_hourly_pay')
plt.yscale('log')
plt.title('Scatter Plot: Total Hours vs Average Hourly Pay (Log Scale on Y)')
plt.show()
```

What becomes clearer once you change the scale?
